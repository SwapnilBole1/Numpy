# Create numpy arrays

```python
import numpy as np
```

```python
#Return evenly spaced numbers over a specified interval.

#Returns `num` evenly spaced samples, calculated over the
#interval [`start`, `stop`].

np.linspace(2,10)
```

```python
# i want 20 numbers between 2-10
np.linspace(2,10,num=12)
```

```python
# i want to know spacing
np.linspace(2,10,num=12,retstep=True)
```

```python
# if we want to exclude stop
np.linspace(2,10,num=12,endpoint=False,retstep=True)
```

```python
p = np.linspace(2,10,num=12,endpoint=False,retstep=True)
p
```

```python
p[-1]
```

```python
np.linspace(2,10,num=12,endpoint=False,retstep=True)[1]
```
